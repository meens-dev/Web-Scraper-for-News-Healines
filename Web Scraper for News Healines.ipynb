```python
import requests
from bs4 import BeautifulSoup

def scrape_news_headlines(url):
    try:
        # Send a GET request to the user-provided website
        response = requests.get(url)
        response.raise_for_status()  # Raise an error if request fails
        
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract headlines (Modify selector based on website structure)
        headlines = []
        for item in soup.find_all('h3'):  # Adjust 'h3' if needed
            headline = item.get_text(strip=True)
            if headline:
                headlines.append(headline)

        return headlines

    except requests.exceptions.RequestException as e:
        print(f"Error fetching data: {e}")
        return []

# Ask user to input the news website URL
news_url = input("Enter the news website URL: ")

# Call the function with the user-provided URL
headlines = scrape_news_headlines(news_url)

# Display the extracted headlines
if headlines:
    print("\nLatest News Headlines:")
    for i, headline in enumerate(headlines, 1):
        print(f"{i}. {headline}")
else:
    print("\nNo headlines found. Try a different URL.")
```
